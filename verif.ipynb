```python
import re
from collections import Counter

# Lire le fichier
with open(r"D:\M2\Langage Naturel 2\TALN\Data\1. Cons√©quence (Co).txt", 'r', encoding='utf-8') as f:
    lignes = f.readlines()

# Compteur pour les types de "de"
types_de = Counter()

print("üîç ANALYSE DES DIFF√âRENTS TYPES DE 'DE'")
print("=" * 50)

for ligne in lignes:
    ligne = ligne.strip()
    if not ligne:
        continue
    
    # Extraire la phrase (ignorer la num√©rotation)
    phrase = re.sub(r'^\d+\.\s*', '', ligne)
    
    # D√©tecter tous les types de "de"
    if re.search(r"\bde l'", phrase):
        types_de["de l'"] += 1
    elif re.search(r'\bdu\b', phrase):  # du = de + le
        types_de["du"] += 1
    elif re.search(r'\bde la\b', phrase):
        types_de["de la"] += 1
    elif re.search(r"\bd'un\b", phrase):
        types_de["d'un"] += 1
    elif re.search(r"\bd'une\b", phrase):
        types_de["d'une"] += 1
    elif re.search(r'\bdes\b', phrase):  # des = de + les
        types_de["des"] += 1
    elif re.search(r'\bde\b', phrase):  # de simple
        types_de["de"] += 1

# Afficher les r√©sultats
print("üìä R√âPARTITION DES TYPES DE 'DE':")
print("=" * 50)

total_de = sum(types_de.values())
for type_de, count in types_de.most_common():
    print(f"{type_de:8} : {count:4} occurrences ({count/total_de*100:.1f}%)")

print("=" * 50)
print(f"üìà Total des 'de' d√©tect√©s : {total_de}")
print(f"üìÑ Nombre total de lignes : {len([l for l in lignes if l.strip()])}")

# Afficher quelques exemples pour chaque type
print("\nüîç EXEMPLES POUR CHAQUE TYPE :")
print("=" * 50)

for type_de in types_de:
    print(f"\n{type_de} :")
    examples = []
    for ligne in lignes:
        ligne = ligne.strip()
        if not ligne:
            continue
        phrase = re.sub(r'^\d+\.\s*', '', ligne)
        
        if type_de == "de l'" and re.search(r"\bde l'", phrase):
            examples.append(phrase)
        elif type_de == "du" and re.search(r'\bdu\b', phrase):
            examples.append(phrase)
        elif type_de == "de la" and re.search(r'\bde la\b', phrase):
            examples.append(phrase)
        elif type_de == "d'un" and re.search(r"\bd'un\b", phrase):
            examples.append(phrase)
        elif type_de == "d'une" and re.search(r"\bd'une\b", phrase):
            examples.append(phrase)
        elif type_de == "des" and re.search(r'\bdes\b', phrase):
            examples.append(phrase)
        elif type_de == "de" and re.search(r'\bde\b', phrase):
            examples.append(phrase)
        
        if len(examples) >= 2:  # Montrer max 2 exemples par type
            break
    
    for ex in examples:
        print(f"   - {ex}")
```

```python
import re

# Lire le fichier
with open(r"D:\M2\Langage Naturel 2\TALN\Data\1. Cons√©quence (Co).txt", 'r', encoding='utf-8') as f:
    lignes = f.readlines()

print("üîç EXEMPLES AVEC PLUS D'UN 'DE'")
print("=" * 50)

compteur_multiple_de = 0

for ligne in lignes:
    ligne = ligne.strip()
    if not ligne:
        continue
    
    # Extraire la phrase (ignorer la num√©rotation)
    phrase = re.sub(r'^\d+\.\s*', '', ligne)
    
    # Compter tous les types de "de"
    de_count = len(re.findall(r'\bde\b|\bdu\b|\bdes\b|\bde l\'|\bd\'un\b|\bd\'une\b', phrase))
    
    if de_count > 1:
        compteur_multiple_de += 1
        print(f"{ligne}")
        print(f"   ‚Üí {de_count} 'de' trouv√©s")

# R√©sultats
total_lignes = len([l for l in lignes if l.strip()])
print("\n" + "=" * 50)
print("üìà R√âSULTATS:")
print(f"‚úÖ Exemples avec 1 'de' : {total_lignes - compteur_multiple_de}")
print(f"‚ö†Ô∏è  Exemples avec 2+ 'de' : {compteur_multiple_de}")
print(f"üìä Total : {total_lignes} exemples")
```
